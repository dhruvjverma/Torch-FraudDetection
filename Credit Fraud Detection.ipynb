```python
from sklearn.preprocessing import StandardScaler
```

```python
# standardizing the Amount column for the Model
sc = StandardScaler()
credit_data["Amount"] = sc.fit_transform(pd.DataFrame(credit_data["Amount"]))
```

```python
credit_data.head()
```

```python
# dropping time column
credit_data = credit_data.drop(["Time"], axis=1)
```

```python
credit_data.head()
```

```python
# check for duplicate values
credit_data.duplicated().any()
```

```python
# removing duplicate data values
credit_data = credit_data.drop_duplicates()
```

```python
credit_data.shape
```

```python
# check for imbalances in data for the target variable
credit_data["Class"].value_counts()
```

```python
# splitting data into dependent and independent variables
X = credit_data.drop(["Class"], axis=1)
y = credit_data["Class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
# initialize all the ML models
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest Classifier": RandomForestClassifier(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
}
```

```python
# training and evaluating metrics for the ml models
for name, clf in classifier.items():
    print(f"\n******************{name}******************")
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    print(f"\n Accuracy: {accuracy_score(y_test, predictions)}")
    print(f"\n Precision: {precision_score(y_test, predictions)}")
    print(f"\n Recall: {recall_score(y_test, predictions)}")
    print(f"\n F1 Score: {f1_score(y_test, predictions)}")
```

```python
# Balancing the data and rerunnning the models
```

```python
# undersampling
normal_data = credit_data[credit_data["Class"] == 0]
fraud_data = credit_data[credit_data["Class"] == 1]
```

```python
normal_data.shape
```

```python
fraud_data.shape
```

```python
normal_data_sample = normal_data.sample(n=473)
```

```python
normal_data_sample.shape
```

```python
# combining normal_data_sample with fraud_data to blance the data set
undersampled_data = pd.concat([normal_data_sample, fraud_data], ignore_index=True)
```

```python
undersampled_data.head()
```

```python
undersampled_data.shape
```

```python
undersampled_data.info()
```

```python
undersampled_data["Class"].value_counts()
```

```python
# splitting data into dependent and independent variables
X1 = undersampled_data.drop(["Class"], axis=1)
y1 = undersampled_data["Class"]
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=5
)
```

```python
X1_train.shape
```

```python
# initialize all the ML models
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest Classifier": RandomForestClassifier(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
}
```

```python
# training and evaluating metrics for the ml models
for name, clf in classifier.items():
    print(f"\n******************{name}******************")
    clf.fit(X1_train, y1_train)
    predictions1 = clf.predict(X1_test)
    print(f"\n Accuracy: {accuracy_score(y1_test, predictions1)}")
    print(f"\n Precision: {precision_score(y1_test, predictions1)}")
    print(f"\n Recall: {recall_score(y1_test, predictions1)}")
    print(f"\n F1 Score: {f1_score(y1_test, predictions1)}")
```
